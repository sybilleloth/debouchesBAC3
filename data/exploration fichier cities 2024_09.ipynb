```python
# Supprimer les colonnes redondantes
cities_df = cities_df.loc[:, ~cities_df.columns.duplicated()]

# Afficher les colonnes restantes pour vérifier
print(cities_df.columns)
```

```python
# Supprimer les lignes identiques
cities_df = cities_df.drop_duplicates()

# Afficher les premières lignes du DataFrame pour vérifier
print(cities_df.head())
cities_df.shape
```

```python
# Supprimer les lignes où 'longitude' et 'latitude' sont identiques à d'autres, en gardant uniquement la première occurrence
cities_df = cities_df.drop_duplicates(subset=['longitude', 'latitude'], keep='first')

# Afficher les premières lignes du DataFrame pour vérifier
print(cities_df.head())
```

```python
cities_df.shape
```

```python
# Conversion de la colonne Académie pour ne garder que des minuscules et remplacer les tirets par des espaces etc avec les accents
esr_df['Académie_modifiée'] = (
    esr_df['Académie']
    .str.lower()
    .str.replace('-', ' ')
    .str.replace('é', 'e')
    .str.replace('è', 'e')
    .str.replace('ç', 'c')
)

# Mapping pour les remplacements
replacement_mapping = {
    'lyon': 'lyon 01',
    'orleans tours': 'orleans',
    'paris': 'paris 01',
    'besançon': 'besancon',
    'nancy metz': 'nancy',
    'aix marseille': 'marseille 01',
    'créteil': 'creteil',
    'normandie': 'rouen',
    'la réunion': 'st denis',
    'guyane': 'cayenne',
    'guadeloupe': 'pointe a pitre',
    'corse': 'corte', # car l'université se trouve à Corte
    'mayotte':'mamoudzou', #lieu de la préfecture 
    'nouvelle caledonie':'noumea', #id
    'polynesie francaise':'papeete' #id
}

# Remplacer les valeurs dans la colonne 'Académie_modifiée'
esr_df['Académie_modifiée'] = esr_df['Académie_modifiée'].replace(replacement_mapping)

# Fusion des DataFrames sur 'Académie_modifiée' et 'city_code'
merged_df = pd.merge(esr_df, cities_df, left_on='Académie_modifiée', right_on='city_code', how='left')

# Sélection des colonnes à conserver
columns_to_keep = [
    'Région', 'Académie', 'Académie_modifiée', 'Etablissement', 
    'Type de diplôme', 'Domaine disciplinaire', 'Discipline', 'Secteur disciplinaire',
    'Libellé du diplôme', 'Nombre de poursuivants', 'Nombre de sortants',
    'Année(s) d\'obtention du diplôme prise(s) en compte', 'Flag', 
    "Taux d\'emploi salarié en France",  "% d'emplois stables parmi les salariés en France", "Mois après la diplomation",
    'latitude', 'longitude', 'department_name', 'region_name', 'region_geojson_name'
]

# Création du DataFrame final avec les colonnes sélectionnées
final_df = merged_df[columns_to_keep]

# Affichage des premières lignes pour vérification
print(final_df.head())
```

```python
#coordonnées sur gps https://www.coordonnees-gps.fr/carte/pays/NC
# nouméa -22.2745264,166.442419 (lat, long)
# papeete -17.5373835,-149.5659964 (lat,long)
# dans dataframe final_df, ajouter les latitude et longitude pour Académie_modifiee = nouvelle caledonie et polynesie francaise
# Ajout des coordonnées GPS pour les académies modifiées "nouvelle caledonie" et "polynesie francaise"
final_df.loc[final_df['Académie_modifiée'] == 'noumea', ['latitude', 'longitude']] = [-22.2745264, 166.442419]
final_df.loc[final_df['Académie_modifiée'] == 'papeete', ['latitude', 'longitude']] = [-17.5373835, -149.5659964]

# Afficher les premières lignes pour vérifier si les coordonnées ont été correctement ajoutées
print(final_df.head(2))
```

```python
# Suppression de la colonne 'Académie modifiée' après traitement
final_df = final_df.drop(columns=['Académie_modifiée'])
```

```python
# Rechercher les lignes contenant 'lifou' pour checker qu'une académie de type lifou qui n'existe pas ne figure pas dans le nouveau dataframe
def test_value(df, value):
    # Rechercher les lignes contenant la valeur spécifiée
    occurrences = df.apply(lambda row: row.astype(str).str.contains(value, case=False, na=False).any(), axis=1)

    # Vérifier s'il y a des occurrences
    if occurrences.any():
        # Sélectionner la première ligne où la valeur apparaît
        first_occurrence_row = df[occurrences].iloc[0]
        return first_occurrence_row
    else:
        return f"La valeur '{value}' n'existe pas dans le DataFrame."

# Appel de la fonction pour tester la présence de 'lifou'
result = test_value(df, 'lifou')

# Afficher le résultat
print(result)
```
